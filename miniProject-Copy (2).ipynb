```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('BKB_WaterQualityData_2020084.csv')
df.head()
```

```python
df.info()
```

```python
df = df[(df['Salinity (ppt)']>0) & (df['Water Temp (?C)']>=0)] #filter data
```

```python
df = df.rename(columns={"Salinity (ppt)": "Salinity", "Water Temp (?C)": "Temp"})
df.head()
```

```python
df.to_csv('clean_data.csv') #store the cleaned data to local computer
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(df['Salinity'], df['Temp'], gridsize=50, cmap='Blues', mincnt=1)
plt.colorbar(label='Amount')
plt.title('Scatter Plot of Salinity vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Salinity (PSU)')
```

```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
from sklearn.metrics import r2_score

# Assuming we have pandas dataframe 'df' with columns 'Salinity' and 'Temp'
salinity = df['Salinity']
temp = df['Temp']

r2_value = r2_score(temp, salinity)
print(r2_value)
```

```python
import scipy.stats

# Assuming we have pandas dataframe 'df' with columns 'Salinity' and 'Temp'
salinity = df['Salinity']
temp = df['Temp']

correlation, p_value = scipy.stats.pearsonr(salinity, temp)
print('Correlation: ', correlation)
print('P-value: ', p_value)
```

```python
%%ai gpt4 -f code
Find the mean, range, median, and mode for both salinity and temp.
```

```python
# Assuming we have pandas dataframe 'df' with columns 'Salinity' and 'Temp'

# Calculate mean for both Salinity and Temp
salinity_mean = df['Salinity'].mean()
temp_mean = df['Temp'].mean()

# Calculate range for both Salinity and Temp
salinity_range = df['Salinity'].max() - df['Salinity'].min()
temp_range = df['Temp'].max() - df['Temp'].min()

# Calculate median for both Salinity and Temp
salinity_median = df['Salinity'].median()
temp_median = df['Temp'].median()

# Calculate mode for both Salinity and Temp
salinity_mode = df['Salinity'].mode()[0]
temp_mode = df['Temp'].mode()[0]

print('Salinity Mean:', salinity_mean)
print('Temp Mean:', temp_mean)

print('Salinity Range:', salinity_range)
print('Temp Range:', temp_range)

print('Salinity Median:', salinity_median)
print('Temp Median:', temp_median)

print('Salinity Mode:', salinity_mode)
print('Temp Mode:', temp_mode)
```
