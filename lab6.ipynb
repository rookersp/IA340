```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is stored in a variable called df
df = pd.DataFrame({
    'fips': ['01', '02', '04', '05', '06', '01', '02', '04', '05', '06'],
    'name': ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California'],
    'income': [36879, 56234, 44282, 34999, 53629, 48952, 80440, 77127, 70176, 92266],
    'year': [2005, 2005, 2005, 2005, 2005, 2019, 2019, 2019, 2019, 2019],
    'income_growth': [None, None, None, None, None, 0.327368, 0.430451, 0.741723, 1.005086, 0.720450]
})

# Plotting the data
df_2019 = df[df['year'] == 2019]
df_2019.set_index('name', inplace=True)
df_2019['income_growth'].plot(kind='bar', figsize=(10,6))
plt.title('Year-over-year income growth rate by state for 2019')
plt.ylabel('Income growth rate')
plt.xlabel('State')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by state and calculate the mean of income_growth
avg_growth = df.groupby('name')['income_growth'].mean().reset_index()

# Plotting
plt.figure(figsize=(10,6))
plt.bar(avg_growth['name'], avg_growth['income_growth'], color='blue')
plt.xlabel('State')
plt.ylabel('Average Income Growth')
plt.title('Average Income Growth by State')
plt.xticks(rotation=90)
plt.show()
```
